```python
data = {
    'BBCA': labels_BBCA,
    'EXCL': labels_EXCL,
    'BBNI': labels_BBNI,
    'BBRI': labels_BBRI,
    'BMRI': labels_BMRI,
    'ASII': labels_ASII,
    'UNVR': labels_UNVR,
    'GGRM': labels_GGRM,
    'INDF': labels_INDF,
    'TLKM': labels_TLKM
}

Data_perubahan = pd.DataFrame(data)
Data_perubahan
```

```python
# Membuat matriks peluang transisi BBCA
states = ['turun', 'naik', 'tetap']
jumlah_state = len(states)
frek_matriks_transisi = np.zeros((jumlah_state, jumlah_state))

for i in range(len(Data_perubahan['BBCA']) - 1):
    current_state = Data_perubahan['BBCA'][i]
    next_state = Data_perubahan['BBCA'][i + 1]
    current_index = states.index(current_state)
    next_index = states.index(next_state)
    frek_matriks_transisi[current_index][next_index] += 1

# Menghitung jumlah frekuensi peralihan state
frekuensi_peralihan_BBCA = np.sum(frek_matriks_transisi, axis=1)

# Normalisasi matriks transisi menjadi peluang transisi
matriks_transisi_BBCA = frek_matriks_transisi / frekuensi_peralihan_BBCA[:, np.newaxis]

# Menampilkan matriks peluang transisi
print("Frekuensi Matriks Peluang Transisi:")
print(frek_matriks_transisi)
print("Jumlah barisan matriks BBCA:")
print(frekuensi_peralihan_BBCA)
print("Matriks Transisi BBCA:")
print(matriks_transisi_BBCA)
```
